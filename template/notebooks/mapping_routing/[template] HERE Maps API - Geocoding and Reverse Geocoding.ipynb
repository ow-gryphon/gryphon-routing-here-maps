# Convert individual addresses

Here we show how to use the functions provided by the 'geocoding' module.

## Get LatLong associated with an address, returning the top result

```python
address = "76 Montague St, Brooklyn, NY, USA"
geocoding_output = geocoding.get_latlong(address, api_key, summarize=True, limit=1)
geocoding_output
```

## Get LatLong associated with an address, returning 5 top results

```python
address = "76 Montague Street"
pd.DataFrame(geocoding.get_latlong(address, api_key, summarize=True, limit=5))
```

## Get address associated with LatLong, returning the top result

latlong should be in the format "latitude,longitude" 

```python
latlong = "40.301,-80.212"
geocoding.get_address(latlong, api_key, summarize=True, limit=1)
```

## Get address associated with LatLong, returning up to the top 5 results

```python
latlong = "40.301,-80.212"
pd.DataFrame(geocoding.get_address(latlong, api_key, summarize=True, limit=5))
```

## Explore addresses near a latlong by category

categories should be a comma-separated list of category-Ids from here: https://developer.here.com/documentation/places/dev_guide/topics/place_categories/places-category-system.html 
names should be a geographic area (country, circular shape, box), see: https://developer.here.com/documentation/geocoding-search-api/api-reference-swagger.html under Browse 

```python
latlong = "40.301,-80.212"
pd.DataFrame(geocoding.browse_address(latlong, api_key, categories=100, summarize=True, limit=5))
```

# Batch geocode by looping through a table with addresses

```python
addresses = pd.read_csv("sample_input/addresses1.csv") # Modify filepath as needed
addresses.head()
```

```python
latlong_df = pd.DataFrame(addresses["Address"].apply(lambda x: geocoding.get_latlong(x, api_key)[0]).tolist())
latlong_df = latlong_df.assign(original_address = addresses['Address'])
latlong_df.head()
```
